```python
d = {}
for i in metadata[metadata['Camps'].isna() == False]['Camps2'].index:
    for j in range(len(metadata['Camps2'][i])):
        if metadata['Camps2'][i][j] in d.keys():
            d[metadata['Camps2'][i][j]] += 1
        else:
            d[metadata['Camps2'][i][j]] = 1
```

I make a copy of the dictionary where I remove the keys with values = 1. I might want to change this threshold at a later point.

```python
d2 = d.copy()
del_list = []
for k, v in d2.items():
    if v == 1:
        del_list.append(k)
```
